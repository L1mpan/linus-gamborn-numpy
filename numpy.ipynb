```python
#3

import numpy as np
import matplotlib.pyplot as plt 

x = np.arange(1,11)
y = 3*x+1

plt.plot(x,y)
plt.ylabel("y")
plt.xlabel("x")
plt.show
print(y)
```

```python
#4

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-2,3)
y = x**2-1

plt.plot(x,y)
plt.ylabel("y")
plt.xlabel("x")
plt.show
print(y)
```

```python
#5

import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,6)
y = -4*t**2+24*t+1

plt.plot(t,y)
plt.ylabel("h(t)")
plt.xlabel("t")
plt.show
```
